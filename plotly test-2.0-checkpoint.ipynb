```python
import dash
import dash_core_components as dcc
import dash_html_components as html
from dash.dependencies import Input, Output
import plotly.graph_objects as go
import pyodbc as pd
import pandas as pan
import numpy as np
import plotly.express as px
from datetime import datetime

server = 'LAPTOP-10GHRH04'
database = 'covid-19-john-hopkins'
username = 'avi'
password = 'admin'

cnxn = pd.connect('DRIVER={ODBC Driver 17 for SQL Server};SERVER=' + server +
                     ';DATABASE=' + database + ';UID=' + username + ';PWD=' + password +
                      ';TRUSTED_CONNECTION=yes;', autocommit=True)
cursor = cnxn.cursor()

months = cursor.execute("""
select distinct cast(month as int) as Month, 
cast(year as int) as Year from [covid-19-john-hopkins].[dbo].[vw-case-per-month] 
order by year asc, month asc;
""").fetchall()

mMonths = {}

for m, y in months :
    dateobj = datetime.strptime(str(m), "%m")
    mName = dateobj.strftime("%B")
    mMonths.update({(int(m), int(y)):mName})

print(mMonths, sep='\n\n')
# mDf = pan.DataFrame(mr,
#                    columns=['Month', 'Year'])
# mDf = mDf.astype({'Month':int, 'Year':int})
#display(mDf)

```

```python
external_stylesheets = ['https://codepen.io/chriddyp/pen/bWLwgP.css']

app = dash.Dash(__name__, external_stylesheets=external_stylesheets)
app.layout = html.Div([
    dcc.Dropdown(
        id='month-dropdown',
        options=[
            {'label': n +' '+str(y), 'value': (v, y)} for (v, y), n in mMonths.items()
#             {'label': 'June 2020', 'value': 6},
#             {'label': 'July 2020', 'value': 7},
#             {'label': 'August 2020', 'value': 8},
#             {'label': 'September 2020', 'value': 9},
#             {'label': 'October 2020', 'value': 10},
#             {'label': 'November 2020', 'value': 11},
#             {'label': 'December 2020', 'value': 12}
        ],
        value=(5, 2020),
    ),
#     dcc.Slider(id='height', min=250, max=700, step=25, value=300,
#                marks={x: str(x) for x in [250, 300, 350, 400, 450, 500, 550, 600, 650, 700]}),
    dcc.Graph(id='graph'),
    html.Div(id='dd-output-container')
])


@app.callback(
    dash.dependencies.Output("graph", "figure"), 
    dash.dependencies.Input('month-dropdown', 'value'))
def draw_graph(value):
    print(value)
    v, y = value
    
    data = cursor.execute("""
    select Country_Region, cast(Total_Confirmed as int) as Total_Confirmed from 
    [covid-19-john-hopkins].[dbo].[vw-case-per-month] where year = """+str(y)+""" and month = """+str(v)+""" 
    order by Total_Confirmed desc;
    """).fetchall()
    
    ar = np.array(data)

    df = pan.DataFrame(ar,
                      columns=['Country Region', 'Total Confirmed'])
    df = df.astype({"Country Region": str, "Total Confirmed": int})
    
    figure=px.bar(df, x = 'Country Region', y = 'Total Confirmed', 
                  title = datetime.strptime(str(v), "%m").strftime("%B")+' 2020', 
             labels={"x": "Country Region", "y": "Total Confirmed"}, height=800)

    return figure
# def resize_figure(height):
#     figure.update_layout(height=int(height))

#     return figure
#    dash.dependencies.Output('dd-output-container', 'children'),
#    [dash.dependencies.Input('demo-dropdown', 'value')])
# def update_output(value):
#     val = value
#     return 'You have selected "{}"'.format(value)

if __name__ == '__main__':
    app.run_server(debug=False)

# app.layout = html.Div([
#     dcc.Dropdown(
#         id='demo-dropdown',
#         options=[
#             {'label': 'May 2020', 'value': 5},
#             {'label': 'June 2020', 'value': 6}
#         ],
#         value=value
#     ),
#     dcc.Graph(
#         id='example-graph',
#         figure=px.bar(df[df['Month'] == value].sort_values(by = ['Total Confirmed'], ascending = False)
#              , x = 'Country Region', y = 'Total Confirmed', title = 'May 2020', 
#              labels={"x": "Country Region", "y": "Total Confirmed"} )
#     ),
#     html.Div(id='dd-output-container')
# ])
    
#fig.show()
```
